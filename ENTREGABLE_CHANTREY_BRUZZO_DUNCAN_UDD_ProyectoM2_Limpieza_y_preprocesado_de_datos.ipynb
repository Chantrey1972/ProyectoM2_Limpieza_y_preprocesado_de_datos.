```python
def eliminar_outliers(df, columnas):
    for col in columnas:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        inf = q1 - 1.5 * iqr
        sup = q3 + 1.5 * iqr

        # Filtramos las filas que están dentro del rango aceptable
        df = df[(df[col] >= inf) & (df[col] <= sup)]

    return df

df_sin_outliers = eliminar_outliers(df, columnas_continua)

df_sin_outliers.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Escalar los datos antes de aplicar PCA
x = df.drop('loan_status', axis=1)  # Eliminar la variable objetivo
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Ajustar PCA con el número de componentes deseado
pca = PCA()
pca.fit(x_scaled)


plt.figure(figsize=(10, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Número de componentes principales')
plt.ylabel('Varianza explicada acumulada')
plt.title('Gráfica de varianza explicada acumulada')
plt.grid(True)
plt.show()


explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)
n_components_optimal = np.argmax(cumulative_variance >= 0.95) + 1

print(f"El número óptimo de componentes principales para explicar al menos el 95% de la varianza es: {n_components_optimal}")

# Ajustar PCA con el número de componentes óptimo
pca = PCA(n_components=n_components_optimal)
x_pca = pca.fit_transform(x_scaled)

```

**Interpretación:**

**Curva creciente:** Al inicio de la gráfica, la varianza explicada acumulada aumenta rápidamente, lo que indica que los primeros componentes principales capturan una gran cantidad de la varianza total del conjunto de datos. Luego, el incremento es más lento conforme se añaden más componentes.

**Punto de inflexión:** Alrededor de los 30 componentes principales, la curva comienza a aplanarse, lo que significa que añadir más componentes adicionales después de este punto no aumenta significativamente la cantidad de varianza explicada. Esto es un indicador de que la mayoría de la información relevante ya está capturada por los primeros componentes.

V**arianza explicada cercana al 100%:** Cerca del final, en el rango de 40-45 componentes, la varianza explicada acumulada llega a 1.0 (100%), lo que significa que todos los componentes principales juntos explican el 100
